```
import random
import string
import operator

population_size = 2000
pw_length = 17

def fittness_password(rand_pw):
  password = 'kaushikDASQmUl78'
  size = len(password)
  count = 0
  dic = {}

  for j in range (0, len(rand_pw)):
    for i in range (0, size):
      if rand_pw[j][i] == password[i]:
        count += 1
      
    dic[rand_pw[j]] = count/size
    count = 0
  return dic


def create_random_password():
  space = string.ascii_letters + '~' + '!' + '@' + '#' + '$' + '%' + '&' + '^' + '*' + string.digits 
  return (''.join(random.choice(space) for j in range (pw_length)))


def selection_by_roulette_wheel(cs_dictionary):
  selection_probability = random.uniform(0,1)
  #print(selection_probability)
  for key, value in cs_dictionary.items():
    if value > selection_probability:
      return key


def crossover(father, mother):
  gene_position = random.randint(0,pw_length-1)
  #print(gene_position)
  return (father[0:gene_position] + mother[gene_position : pw_length])

def mutation(child_gene):
  gene_position = random.randint(0,pw_length-1)
  space = string.ascii_letters + '~' + '!' + '@' + '#' + '$' + '%' + '&' + '^' + '*' + string.digits
  dna = random.choice(space)

  mutated_gene = child_gene[0 : gene_position] + dna + child_gene[gene_position+1 : pw_length]
  return(mutated_gene)

def create_population(n):
  tot_fitness = 0
  password_list  =[]
  for i in range (n):
    password_list.append(create_random_password())
  #print(password_list)
  fitness_dict = fittness_password(password_list)

  for k,v in fitness_dict.items():
    tot_fitness += v 

  for k,v in fitness_dict.items():
    fitness_dict[k] = v / tot_fitness 


  sorted_fitness_dict = dict(sorted(fitness_dict.items(), key=operator.itemgetter(1)))
  print(sorted_fitness_dict)
  previous_val = 0
  for key, val in sorted_fitness_dict.items():
    sorted_fitness_dict[key] += previous_val
    previous_val = sorted_fitness_dict[key]
  new_sorted_fitness_dict = sorted_fitness_dict

  while (1):
    new_password_list = []
    new_fitness_dict = {}
    
    for i in range (population_size):

      father = selection_by_roulette_wheel(new_sorted_fitness_dict)
      mother = selection_by_roulette_wheel(new_sorted_fitness_dict)

      #print(father,mother)

      child = crossover(father, mother)

      chance = random.uniform(0,1)
      if chance <0.05:
        child = mutation(child)

      new_password_list.append(child)

    print('***********')
    new_fitness_dict = fittness_password(new_password_list)
    print('Maximum fitness is {} and total elements are {}'.format(max(new_fitness_dict.items(), key=operator.itemgetter(1))[1], len(new_fitness_dict)))
    print('Maximum fit password is {}'.format(max(new_fitness_dict.items(), key=operator.itemgetter(1))[0]))

    if (max(new_fitness_dict.items(), key=operator.itemgetter(1))[1]) ==1:
      print('desired password is {}'.format(max(new_fitness_dict.items(), key=operator.itemgetter(1))[0]))
      #print(new_fitness_dict)
      break

    tot_fitness = 0
    for k,v in new_fitness_dict.items():
      tot_fitness += v 

    for k,v in new_fitness_dict.items():
      new_fitness_dict[k] = v / tot_fitness 

    #print(new_fitness_dict)
    new_sorted_fitness_dict = dict(sorted(new_fitness_dict.items(), key=operator.itemgetter(1)))
    

    previous_val = 0
    for key, val in new_sorted_fitness_dict.items():
      new_sorted_fitness_dict[key] += previous_val
      previous_val = new_sorted_fitness_dict[key]
    #print(new_sorted_fitness_dict)
    

create_population(population_size)
```
